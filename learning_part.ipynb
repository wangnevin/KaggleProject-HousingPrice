```python
%tensorflow_version 1.x
import tensorflow as tf 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw = pd.read_csv('train.csv')
data = pd.read_csv('clean_data.csv')
print("The shape of data: {}".format(data.shape))
data.head()
```

```python
def generate_X(data):
  X = data.drop('Id',axis=1)
  X = X.values

  return X

def generate_y_train(data):
  y_train = data['SalePrice'].values
  return y_train

def split_X_train_test(data, train_num):
  X_train = X[:train_num,:]
  X_test = X[train_num:,:]
  return X_train, X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

X = generate_X(data)
y_train = generate_y_train(raw)

print("The shape of X: {}".format(X.shape))
print("The shape of y_train: {}".format(y_train.shape))
```

```python
X = MinMaxScaler().fit_transform(X)
pca = PCA(n_components=0.97)
X = pca.fit_transform(X)
print("The shape of X after PCA: {}".format(X.shape))
print("The shape of y_train: {}".format(y_train.shape))
```

```python
X_train, X_test = split_X_train_test(X, 1460)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import Lasso

la = Lasso()

la.fit(X_train,y_train)
print(la.score(X_train,y_train))
```

```python
result = la.predict(X_test)
print(result.shape)
print(result)

Id_s = pd.Series(np.arange(1461,2920))
result_s = pd.Series(result)

result_df = pd.DataFrame()
result_df['Id'] = Id_s
result_df['SalePrice'] = result_s

print(result_df.shape)
print(result_df.head())

result_df.to_csv('submission2.csv', index=False)
```
